```python
import pandas as pd
import seaborn as sns

# Load a sample dataset
df = sns.load_dataset('titanic')

# Display the first 5 rows
print("Original DataFrame:")
display(df.head())

# Perform some data manipulations
# Drop the 'deck' column as it has many missing values
df = df.drop('deck', axis=1)

# Fill missing values in 'age' with the median
df['age'] = df['age'].fillna(df['age'].median())

# Fill missing values in 'embarked' with the most frequent value
df['embarked'] = df['embarked'].fillna(df['embarked'].mode()[0])

# Perform some filtering
# Filter for passengers who survived
survived_df = df[df['survived'] == 1]

# Filter for female passengers in the first class
female_first_class_df = df[(df['sex'] == 'female') & (df['pclass'] == 1)]

print("\nDataFrame after dropping 'deck' and filling missing values:")
display(df.head())

print("\nDataFrame of survived passengers:")
display(survived_df.head())

print("\nDataFrame of female passengers in first class:")
display(female_first_class_df.head())
```
